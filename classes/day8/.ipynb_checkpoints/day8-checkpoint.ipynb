```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy

def plot_image(image, title=None):
    if len(image.shape) == 3 and image.shape[-1] >= 3:
        plt.imshow(image)
    else:
        plt.imshow(image, cmap="gray")
    
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
img = cv2.imread("indian_flag.jpg", cv2.IMREAD_GRAYSCALE)
print("Image dimensions =", img.shape)
plot_image(img, "original image")
u, d, vt = np.linalg.svd(img)
```

```python
components = 5

d_lossy = deepcopy(d)
d_lossy[d < np.unique(d)[-components]] = 0

diagonal = np.zeros(img.shape)
diagonal[:u.shape[0], :u.shape[1]] = np.diag(d_lossy)

m = u @ diagonal @ vt
plot_image(m, "compressed image")
```

```python
compressed_size = components*(u.shape[1] + vt.shape[0] + 1)
original_size = u.shape[1]*vt.shape[0] 
print("Compression Ratio: %.3f" % (original_size/compressed_size))
```

```python
def compress_image(components, img):
    u, d, vt = np.linalg.svd(img)
    d_lossy = deepcopy(d)
    d_lossy[d < np.unique(d)[-components]] = 0

    diagonal = np.zeros(img.shape)
    diagonal[:u.shape[0], :u.shape[1]] = np.diag(d_lossy)

    m = u @ diagonal @ vt
    compressed_size = components*(u.shape[1] + vt.shape[0] + 1)
    original_size = u.shape[1]*vt.shape[0]
    return m, (original_size/compressed_size)

num_of_components = 50
plot_image(img, "original image")
compressed_img, compression_ratio = compress_image(num_of_components, img)
plot_image(compressed_img, "compressed image - not noticable difference")
print("Compression Ratio: %.3f" % compression_ratio)
print("Number of components:", num_of_components)
```

# Unsupervised Learning

```python
import matplotlib.cm as cm
from sklearn.metrics.pairwise import euclidean_distances

def plot_clusters(points, cluster_labels, show_annotation=True, offset=0.05, annotations=None):
    clusters = np.array(cluster_labels)
    clusters_unique = np.unique(clusters)
    for cluster in clusters_unique:
        plt.scatter(points[:,0][clusters == cluster], points[:,1][clusters == cluster], label='class '+str(cluster))
    
    if show_annotation:
        for i in range(len(points)):
            if annotations is None:
                plt.annotate(str(i), xy=points[i], xytext=points[i]+offset)
            else:
                plt.annotate(str(annotations[i]), xy=points[i], xytext=points[i]+offset)
    plt.legend()
```

```python
np.random.seed(2)
points = np.vstack((np.random.normal(0,0.5,size=(4,2)),np.random.normal(3,0.5,size=(4,2))))
points = np.vstack((points, np.array([[1.5,1.5]])))
points = np.vstack((points, np.array([[1.25,1.25]])))

cluster_labels = [0 for i in range(10)]
plot_clusters(points, cluster_labels)
plt.show()
```

```python
def heatmap(grid, xlabels, ylabels, xrot=90):   
    cmap = cm.get_cmap("viridis")
    plt.imshow(grid, cmap=cmap)
    plt.xticks(np.arange(0,grid.shape[1],step=1), xlabels, rotation=xrot)
    plt.yticks(np.arange(0, grid.shape[0], step=1), ylabels)
```

```python
grid = euclidean_distances(points, points)
grid = (grid - grid.min())/(grid.max() - grid.min())
xlabels = np.arange(0, len(points))
ylabels = np.arange(0, len(points))
heatmap(grid, xlabels, ylabels, xrot=0)
```

```python
def plot_cluster_centers(centers, marker='x', offset=0.05):
    plt.scatter(centers[:,0], centers[:,1], marker=marker)
    
    for i in range(len(centers)):
        plt.annotate(str(i), xy=centers[i], xytext=centers[i]+offset)
```

```python
def initialize(k, points):
    new_points = deepcopy(points)
    np.random.shuffle(new_points)
    return new_points[:k]

def assign(points, centers):
    return euclidean_distances(points, centers).argmin(axis=1)

def recalculate_clusters(points, cluster_labels, k):
    new_centers = np.zeros((k, 2))
    for i in range(k):
        new_centers[i] = points[cluster_labels == i].mean(axis=0)
    return new_centers 
        
def kmeans(points, k):
    centers = initialize(k, points)
    last_centers = np.zeros_like(centers)
    cluster_labels = np.zeros_like(points)

    while not (centers == last_centers).all():
        last_centers = deepcopy(centers)
        cluster_labels = assign(points, centers)
        centers = recalculate_clusters(points, cluster_labels, len(centers))
    
    return centers, cluster_labels
```

```python
centers, cluster_labels = kmeans(points, 3)
plot_clusters(points, cluster_labels)
plot_cluster_centers(centers)
plt.show()
```

```python
import numpy as np

def cart2pol(points):
    x, y = points[:,0], points[:,1]
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    return np.array([r, theta]).T

def pol2cart(r, theta):
    r, theta = points[:,0], points[:,1]
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return np.array([x, y]).T

def only_radial(points):
    x, y = points[:,0], points[:,1]
    r = np.sqrt(x**2 + y**2)
    return np.array([r, r]).T
    
centers, cluster_labels = kmeans(only_radial(points), 2)
plot_clusters(points, cluster_labels, show_annotation=False)
# plot_cluster_centers(centers)
plt.show()
```

```python
import os
import json

data_path = 'task3/Final Dataset/'

names = []
genre_freq = []
for i in os.listdir(data_path):
    with open(os.path.join(data_path, i), 'r') as f:
        data_dict = json.load(f)
    # Akshat seems to have a wierdly large and not so interpretable data
    if data_dict['username'] == "Akshat":
        continue
    names.append(data_dict['username'])
    genre_freq.append(data_dict['genre_freq_divided'])

print(names)
```

```python
chosen_genres = list(genre_freq[0].keys())[:2]
print(chosen_genres)
points = np.zeros((len(names), len(chosen_genres)))
for i, name in enumerate(names):
    for g, genre in enumerate(chosen_genres):
        points[i][g] = genre_freq[i][genre]

print(points)
```

```python
centers, cluster_labels = kmeans(points, 3)
plot_clusters(points, cluster_labels, annotations=names)
plt.xlabel(chosen_genres[0])
plt.ylabel(chosen_genres[1])
plt.show()
```

```python
def cluster_on_genres(chosen_genres, k):
    # Only 2D graphs look good
    chosen_genres = chosen_genres[:2]
    points = np.zeros((len(names), len(chosen_genres)))
    for i, name in enumerate(names):
        for g, genre in enumerate(chosen_genres):
            points[i][g] = genre_freq[i][genre]

    centers, cluster_labels = kmeans(points, k)
    plot_clusters(points, cluster_labels, annotations=names)
    plt.xlabel(chosen_genres[0])
    plt.ylabel(chosen_genres[1])

cluster_on_genres(['History', 'Documentary'], 2)
plt.show()
```
