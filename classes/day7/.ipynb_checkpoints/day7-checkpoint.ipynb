```python
import numpy as np

threshold = -0.2
lr = 0.1

def gate_data(gate='and'):
	X = np.float32([
		[0,0],
		[0,1],
		[1,0],
		[1,1]
	])

	if gate == "and":
		Y = np.float32([
			0,0,0,1
		])

	if gate == "or":
		Y = np.float32([
			0,1,1,1
		])

	if gate == "xor":
		Y = np.float32([
			0,1,1,0
		])
	
    
	return X, Y

def activation(value):
    return int(value + threshold > 0)

def loss(y_pred, y):
    return np.sum(abs(y - y_pred))

n_epochs = 100

gates = ["and", "or", "xor"]

for gate in gates[:2]:
	X, Y = gate_data(gate)
	W = np.random.uniform(low=-1,high=1,size=(2))

	print()
	print(gate.upper(), "gate:")
    
	Y_pred = np.zeros_like(Y)
	error = np.zeros_like(Y)
	np.set_printoptions(precision=3, suppress=True)

	for epoch in range(1,n_epochs+1):
		for i, x in enumerate(X):
			Y_pred[i] = activation(np.sum(x*W))
			error[i] = Y[i] - Y_pred[i]
			W += lr*(error[i])*x

		print("## Epoch:", epoch,"\t", W, error)
		
		if loss(Y_pred, Y) == 0:
			break
	
	print("Final weights:", W)
```
