```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from copy import deepcopy

def plot_image(image):
    plt.imshow(image, cmap=plt.cm.binary)
    plt.show()
    
def histo2(image):
    plt.hist(image.ravel(), bins=256, range=(0, 256), fc='k', ec='k')
    plt.show()

def mean_threshold(image, threshold=None):
    image = deepcopy(image)
    if threshold is None:
        threshold = image.mean()
    print("Threshold:", int(threshold))
    image[image >= threshold] = 255
    image[image < threshold] = 0
    return image

def hw(image):
    return image.shape[:2]
```

```python
import glob
import os

orb = cv2.ORB_create()
sift =cv2.xfeatures2d.SIFT_create()
```

```python
def trim(frame):
          #crop top
          if not np.sum(frame[0]):
              return trim(frame[1:])
          #crop bottom
          if not np.sum(frame[-1]):
              return trim(frame[:-2])
          #crop left
          if not np.sum(frame[:,0]):
              return trim(frame[:,1:])
          #crop right
          if not np.sum(frame[:,-1]):
              return trim(frame[:,:-2])
          return frame

def panaroma(image2, image1):

    image1_gray=cv2.cvtColor(image1,cv2.COLOR_RGB2GRAY)
    image2_gray=cv2.cvtColor(image2,cv2.COLOR_RGB2GRAY)

    image1 = deepcopy(image1)
    image2 = deepcopy(image2)

    kp1, des1 = orb.detectAndCompute(image1_gray,None)
    kp2, des2 = orb.detectAndCompute(image2_gray,None)

#     fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,8), constrained_layout=False)
#     ax1.imshow(cv2.drawKeypoints(image1_gray,kp1,None,color=(0,255,0)))
#     ax1.set_xlabel("(a)", fontsize=14)
#     ax2.imshow(cv2.drawKeypoints(image2_gray,kp2,None,color=(0,255,0)))
#     ax2.set_xlabel("(b)", fontsize=14)
#     plt.show()
    
    match = cv2.BFMatcher()
    matches = match.knnMatch(des1,des2,k=2)
    good = []
    for m,n in matches:
        if m.distance < 0.03*n.distance:
            good.append(m)

    draw_params = dict(matchColor=(0,255,0),
                           singlePointColor=None,
                           flags=2)

    img3 = cv2.drawMatches(image2_gray,kp2,image1_gray,kp1,good,None,**draw_params)
    plot_image(img3)

    MIN_MATCH_COUNT = 10
    if len(good) > MIN_MATCH_COUNT:
        src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
        dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)

        M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)

        h,w = image2_gray.shape
        pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
        dst = cv2.perspectiveTransform(pts, M)
        dst = cv2.warpPerspective(image1,M,(image2.shape[1] + image1.shape[1], image2.shape[0] + image1.shape[0]))
        dst[0:image2.shape[0],0:image2.shape[1]] = image2
        return trim(dst)
    else:
        print("Not enought matches are found - {}/{}".format(len(good), MIN_MATCH_COUNT))
        return np.zeros_like(image1)

def get_all_images(images_path):
    images = []
    for i in os.listdir(images_path):
        images.append( cv2.cvtColor( cv2.imread(os.path.join(images_path, i)), cv2.COLOR_BGR2RGB ) )
    return images

def multiple_panaroma(images):
    composed_image = images[0]
    for i in range(1,len(images)):
        composed_image = panaroma(composed_image, images[1])
    return composed_image

plot_image(multiple_panaroma(get_all_images('./Panorama/Eg1/')))
```

```python
tsukaba_l = cv2.imread("tsukuba_l.png", cv2.IMREAD_GRAYSCALE)
tsukaba_r = cv2.imread("tsukuba_r.png", cv2.IMREAD_GRAYSCALE)

stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(tsukaba_l, tsukaba_r)

disparity_threshold = 120

tsukaba_new = deepcopy(tsukaba_l)
tsukaba_new[disparity < disparity_threshold] = 0
plot_image(255 - tsukaba_new)
```

```python
view0 = cv2.imread("view0.png", cv2.IMREAD_COLOR)
view1 = cv2.imread("view1.png", cv2.IMREAD_COLOR)

view0_gray=cv2.cvtColor(view0,cv2.COLOR_BGR2GRAY)
view1_gray=cv2.cvtColor(view1,cv2.COLOR_BGR2GRAY)

view0=cv2.cvtColor(view0,cv2.COLOR_BGR2RGB)
view1=cv2.cvtColor(view1,cv2.COLOR_BGR2RGB)

def ssd_disp_calc(image_l, image_r):
    patch_size = 11
    patch_side = int(patch_size/2)
    h, w = image_l.shape[:2]
    disp = np.zeros_like(image_l).astype("int")
    matched = np.zeros((w), dtype="int")

    for i in range(h):
        for j in range(patch_side, w-patch_side):
            left_patch = image_l[i, j-patch_side:j+patch_side]
            patch_diff = np.zeros((w))

            for k in range(patch_side, w-patch_side):
                right_patch = image_r[i, k-patch_side:k+patch_side]
                patch_diff[k] = np.sum((right_patch - left_patch)**2)

            matched[j] = np.argmin(patch_diff[patch_side: w-patch_side])
            disp[i][j] = (j - matched[j])

    return disp

disp = ssd_disp_calc(view0_gray, view1_gray)
```

```python
image_new = deepcopy(view0)
image_new[disp <= 134] = 0

def fit_as_image(image):
    image = deepcopy(image)
    
    mini = image.min()
    maxi = image.max()

    image = (image - mini)/(mini - maxi)
    return (image*255).astype("uint8")

plot_image(fit_as_image(disp))
plot_image(view0)
```
