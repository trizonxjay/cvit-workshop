tanh function:$$
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

Gaussian Function:$$
f(x) = e^{-x^2}
$$

Rectified Linear Unit(ReLU):$$
f(x) = max(0, x)
$$

Leaky ReLU:$$
f(x) = max(0.1x, x)
$$

Step Function $$ f(x) = 1 \forall x > 0, otherwise 0 $$

Natural Logarithm:$$
ln(x) = log_e(x)
$$

Exponential function:$$
f(x) = e^x
$$

Relation between tanh and sigmoid
$$
tanh(x) = 2\sigma(2x) -1
$$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return max(0,x)

def leaky_relu(x):
    return max(0.1*x,x)

def gaussian(x):
    return np.exp(-x**2)

def step(x):
    return 1 if x>0 else 0

def tanh(x):
    return (np.exp(2*x)-1)/(np.exp(2*x)+1)
```

```python
%matplotlib inline

X = np.arange(-20,20)

plt.figure()
plt.plot(X,np.vectorize(sigmoid)(0.5*X))
plt.title('Sigmoid')
plt.show()

plt.figure()
plt.plot(X,np.vectorize(relu)(X))
plt.title('ReLU')
plt.show()

plt.figure()
plt.plot(X,np.vectorize(leaky_relu)(X))
plt.title('Leaky ReLU')
plt.show()

plt.figure()
X = np.arange(-20,20,0.01)
plt.plot(X,np.vectorize(gaussian)(X))
plt.title('Gaussian')
plt.show()

plt.figure()
plt.plot(X,np.vectorize(step)(X))
plt.title('Step')
plt.show()

plt.figure()
plt.plot(X,np.vectorize(tanh)(X))
plt.title('Tanh')
plt.show()

plt.figure()
plt.plot(X,np.vectorize(np.exp)(X))
plt.title('Exponential')
plt.show()

plt.figure()
X = np.arange(0.01,20,0.01)
plt.plot(X,np.vectorize(np.log)(X))
plt.title('Log')
plt.show()
```
